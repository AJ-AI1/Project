```python
# Dependencies
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
os.getcwd()
```

```python
# Take in the CSV data and read it into pandas
rates ='..\Project 1\Resources\Ratesmotor.csv'
rates_df = pd.read_csv(rates)
rates_df.head()
```

```python
# Show mulitple specific columns needed
rates_sorted_df = rates_df[["state", "rate", "fuel_type", "abbrev", "MMFR_year", "effective_date"]]
rates_sorted_df.head()
```

```python
#Set the abbrev column as the index
rates_sorted_df.set_index('abbrev', inplace=True)
rates_sorted_df
```

```python
#Filter the data by year so that only the rows for the year 2023 are in a DataFrame
rates_years_df = rates_sorted_df[rates_sorted_df['MMFR_year'] == 2023]
rates_years_df
```

```python
#Sort the data by the fuel_type column to show all the rows for Diesel
rates_years_df_sorted = rates_years_df[rates_years_df['fuel_type'] == 'Diesel']
rates_years_df_sorted.head()
```

```python
#sort the data by rate in descending order
rates_diesel_df = rates_years_df_sorted.sort_values(by='rate', ascending=False)
rates_diesel_df.head()
```

```python
#show the last 5 rows of the data
rates_diesel_df.tail()
```

```python
#set the x_axis to be the length of the data
x_axis = np.arange(len(rates_diesel_df))
tick_locations = [value for value in x_axis]
```

```python
# Plot a bar chart and create a list indicating where to write x labels and set figure size to adjust for space  
x = rates_diesel_df["state"]
y = rates_diesel_df["rate"]

plt.figure(figsize=(8, 4))
plt.bar(x, y, color='blue')

# Customize the chart
plt.title("Average Diesel Tax Rates per State in 2023")
plt.xlabel("State")
plt.ylabel("Diesel Tax Rates")
plt.xticks(rotation=90)

# Show the chart
plt.show()
```

```python
# Create a group based on the values in the 'effective_date' column
type = rates_years_df_sorted.groupby('effective_date')

# Count how many times each effective_date appears
rate_types = type['effective_date'].count()

rate_types.head()
```

```python
#Set the xlabel and ylabel and plot the pie chart
count_chart = rate_types.plot(kind='pie', figsize=(8,6))

count_chart.set_xlabel("Effective Date")
count_chart.set_ylabel("Rates")
count_chart.set_title("Diesel Tax Rates per Day from 2013 - 2023")

plt.show()
```
